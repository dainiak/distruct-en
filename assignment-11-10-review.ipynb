### Task 4.
Let $S=\{0,\,1,\,2,\,3\}$. Use `product` function to build up a set $S^3$. Use this set to answer a question: how many integers can be represented as sum of squares of at most three numbers from $S$.

```python
# importing the library
from itertools import product

# The line below is equivalent to S = {0, 1, 2, 3}
S = list(range(4))

S3 = set(product(S, repeat=3))
# print(S3)

sums_of_squares = {i**2 + j**2 + k**2 for (i, j, k) in S3}

print( len(sums_of_squares) )
```
